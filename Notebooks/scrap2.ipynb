```julia
include("circuits.jl")
using Plots
```

```julia
#try to see if DRT can track changes with randles circuits
randles_circuits = [0.23r-(10r-0.25*ws^b)/.1q for b in 60:10:100]
randles_fits = [compute_drt(circuit.ω,circuit.Z;showplot=false) for circuit in randles_circuits]
```

```julia
τ = randles_fits[1]["τ"]
colors = palette([:purple, :orange],length(randles_fits))
plt = plot()
for (fit,color) in zip(randles_fits,colors)
    plot!(plt,τ,fit["drt"];color=color, xaxis = :log,legend = :topleft)
end
display(plt)
```
